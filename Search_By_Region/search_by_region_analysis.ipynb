```python
import pandas as pd

with open('data/google.txt', 'r', encoding='utf-8') as fp:
    data = fp.read().splitlines()

region_dict = {}

with open('region.txt', 'r', encoding='utf-8') as fp:
    regions = fp.read().splitlines()
    for region in regions:
        region_dict[region] = []

for line in data:
    term, link, kwd, pagenum, time = eval(line)
    region_dict[kwd].append((term, link, kwd, pagenum, time))

region_with_promotions_in_page1 = []
region_with_promotions_in_page2 = []
region_with_promotions_in_page5 = []
region_with_promotions_in_page10 = []
region_with_promotions_in_all_page = []
for region in region_dict.keys():
    for item in region_dict[region]:
        if item[3] == 0:
            region_with_promotions_in_page1.append(region)
            region_with_promotions_in_page2.append(region)
            region_with_promotions_in_page5.append(region)
            region_with_promotions_in_page10.append(region)
            region_with_promotions_in_all_page.append(region)
            break
        if item[3] == 1:
            region_with_promotions_in_page2.append(region)
            region_with_promotions_in_page5.append(region)
            region_with_promotions_in_page10.append(region)
            region_with_promotions_in_all_page.append(region)
            break
        if item[3] > 1 and item[3] <= 4:
            region_with_promotions_in_page5.append(region)
            region_with_promotions_in_page10.append(region)
            region_with_promotions_in_all_page.append(region)
            break
        if item[3] > 4 and item[3] <= 9:
            region_with_promotions_in_page10.append(region)
            region_with_promotions_in_all_page.append(region)
            break
        if item[3] > 9:
            region_with_promotions_in_all_page.append(region)
            break

num_of_promotions_in_page1 = 0
num_of_promotions_in_page2 = 0
num_of_promotions_in_page5 = 0
num_of_promotions_in_page10 = 0
num_of_promotions_in_all_page = 0
for region in region_dict.keys():
    for item in region_dict[region]:
        if item[3] == 0:
            num_of_promotions_in_page1 += 1
            num_of_promotions_in_page2 += 1
            num_of_promotions_in_page5 += 1
            num_of_promotions_in_page10 += 1
            num_of_promotions_in_all_page += 1
        if item[3] == 1:
            num_of_promotions_in_page2 += 1
            num_of_promotions_in_page5 += 1
            num_of_promotions_in_page10 += 1
            num_of_promotions_in_all_page += 1
        if item[3] > 1 and item[3] <= 4:
            num_of_promotions_in_page5 += 1
            num_of_promotions_in_page10 += 1
            num_of_promotions_in_all_page += 1
        if item[3] > 4 and item[3] <= 9:
            num_of_promotions_in_page10 += 1
            num_of_promotions_in_all_page += 1
        if item[3] > 9:
            num_of_promotions_in_all_page += 1

table_data = {'page top 1':[], 'page top 2':[], 'page top 5':[], 'page top 10':[]}
table_data['page top 1'] = [len(region_with_promotions_in_page1), len(region_with_promotions_in_page1) / len(regions), num_of_promotions_in_page1]
table_data['page top 2'] = [len(region_with_promotions_in_page2), len(region_with_promotions_in_page2) / len(regions), num_of_promotions_in_page2]
table_data['page top 5'] = [len(region_with_promotions_in_page5), len(region_with_promotions_in_page5) / len(regions), num_of_promotions_in_page5]
table_data['page top 10'] = [len(region_with_promotions_in_page10), len(region_with_promotions_in_page10) / len(regions), num_of_promotions_in_page10]
table_data['all page'] = [len(region_with_promotions_in_all_page), len(region_with_promotions_in_all_page) / len(regions), num_of_promotions_in_all_page]

print(f'num of all region keywords = {len(regions)}, result of Google:')

index = ['num of region keywords with cybercrime result in search page', 'rate', 'num of cybercrime promotions']
pd.DataFrame(data=table_data, index = index).T
```
